#### Import json files, parse out, characteristics of each post, and join together in a dataframe

```python


def find_caption(node):
    try:
        node_caption = node['edge_media_to_caption']['edges'][0]['node']['text']
    except:
        node_caption = ''
    return node_caption


class post:
    def __init__(self, node):
        self.id = int(node['id'])
        self.likes = int(node['edge_media_preview_like']['count'])
        self.comments = int(node['edge_media_to_comment']['count'])
        self.timestamp = datetime.fromtimestamp(int(node['taken_at_timestamp']))
        self.caption = find_caption(node)
        self.hashtags = re.findall(r'(?<=#)\w+', self.caption)
        self.pic_dim = node['dimensions']
        self.pic_url = node['display_url']
        
        
path = '/Users/acorn_leaf/Documents/cattown/data/raw/'

cattownposts_raw = pd.DataFrame()
for file in glob.glob(os.path.join(path, '*.json.xz')):
    bytes_data = lzma.open(file).read()
    json_str = bytes_data.decode('utf-8')
    post_info = json.loads(json_str)
    if post_info['instaloader']['node_type'] == 'Post':
        node_info = post_info['node']
        post_dict = post(node_info).__dict__
        post_df = pd.DataFrame([post_dict])
        cattownposts_raw = pd.concat([cattownposts_raw,post_df], axis=0)
cattownposts_raw.reset_index(inplace=True, drop=True)
cattownposts_raw['date'] = cattownposts_raw['timestamp'].dt.date

cattownposts_raw
```

#### Clean up the hashtags

```python
cattownposts = cattownposts_raw

# Can I use re to remove all punctuation?
drop_list = [' ',',','.','?!','...',':','!','?','(',')','\\ud83d','\\udc3e','\\u00e9e','\\udc08','\\u2019s']

# # There's probably a module that deals with plurals/lemmatization
# replacements = {'cats':'cat','days','day','bday':'birthday'}

# Make everything lower case
def lowercase(val):
    val = [v.lower() for v in val]
    return val
cattownposts['hashtags'] = cattownposts['hashtags'].apply(lowercase)

# Drop and make replacements
for word in drop_list:
    cattownposts['hashtags'] = [[item.replace(word, '') for item in lst] for lst in cattownposts['hashtags']]
# for key,value in replacements.items():
#     cattownposts['hashtags'] = [[item.replace(key,value) for item in lst] for lst in cattownposts['hashtags']]  

cattownposts
```

#### Count number of likes and comments per post over time

```python
# First create a deep copy of cattownposts_clean so
# that we don't update/change the original dataframe
by_post = cattownposts.copy()
by_post.index = pd.to_datetime(by_post['date'])
by_post = by_post[['id','likes','comments']]


def create_counts_dataframe(time_unit='1w'):
    posts = by_post[['id']].resample(time_unit).count()
    likes = by_post[['likes','comments']].resample(time_unit).sum()
    count_df = likes.merge(posts,how='left',left_index=True,right_index=True)
    count_df['likes_ave'] = count_df['likes']/count_df['id']
    count_df['comments_ave'] = count_df['comments']/count_df['id']
    count_df.rename({'id':'number_of_posts'}, axis=1, inplace=True)
    return count_df

# daily = create_counts_dataframe('1d')
weekly = create_counts_dataframe('1w')
monthly = create_counts_dataframe('1m')

weekly.head()
```

#### Plot_likes_comments

```python


def plot_likes_comments(df,
                        average=True,
                        start_date=None,
                        end_date=None,
                        save=False):
    """Plots number of likes and comments  by date.
    
    Variable df can be the original data (number of likes and
    comments per post) or can be a summary by day, week, month, etc."""
    
    fig = plt.figure(figsize=(8, 8))
    
    if average:
        y_likes = df['likes']/df['number_of_posts']
        y_comments = df['comments']/df['number_of_posts']
        title = 'Average number of likes and comments'
    else:
        y_likes = df['likes']
        y_comments = df['comments']
        title = 'Number of likes and comments'
        
    plt.title(title)
    
    # Plot likes
    likes = fig.add_subplot(1,1,1)
    likes.scatter(x=df.index,
                  y=y_likes,
                  label='Likes',
                  color='#1a7d93',
                  alpha=0.8)
    likes.set_xlabel('Date')
#     myFmt = mdates.DateFormatter('%m-%Y')
#     likes.xaxis.set_major_formatter(myFmt)
    likes.set_ylabel('Likes', color='#1a7d93')
    likes.tick_params(axis='y', colors='#1a7d93')

    # Plot comments
    comments = likes.twinx()
    comments.scatter(x=df.index,
                     y=y_comments,
                     label='Comments',
                     color='orange',
                     alpha=0.5)
    comments.set_ylabel('Comments', color='orange')
    comments.tick_params(axis='y', colors='orange')

    if save:
        plt.savefig(r'/Users/acorn_leaf/Documents/cattown/data/processed/cattown_counts.png')
```

```python
matplotlib.rcParams.update({'font.size': 14})

plot_likes_comments(weekly, save=True)
```

#### Count average number of likes and comments per hashtag

```python
hashtags = list(set([item for sublist in cattownposts['hashtags'].tolist() for item in sublist]))

hashtags_likes = []
hashtags_comments = []
hashtags_counts = []
for ht in hashtags:
    hashtags_likes.append(cattownposts[[ht in x for x in cattownposts['hashtags']]]['likes'].sum())
    hashtags_comments.append(cattownposts[[ht in x for x in cattownposts['hashtags']]]['comments'].sum())
    hashtags_counts.append(cattownposts[[ht in x for x in cattownposts['hashtags']]]['id'].count())
hashtags_stats = pd.DataFrame({'hashtag': hashtags,
                               'likes': hashtags_likes,
                               'comments': hashtags_comments,
                               'counts': hashtags_counts})
hashtags_stats['ave_likes'] = hashtags_stats['likes']/hashtags_stats['counts']
hashtags_stats['ave_comments'] = hashtags_stats['comments']/hashtags_stats['counts']
hashtags_stats.sort_values(['ave_likes', 'ave_comments'], inplace=True, ascending=False)
hashtags_stats.head()
```

#### Finds the hastag with the most posts and lists those posts

```python
max_counts = hashtags_stats['counts'].max()
maxed_out = hashtags_stats[hashtags_stats['counts'] == max_counts].reset_index(drop=True)
print(len(maxed_out))
print(maxed_out)

hashtag_posts = dict.fromkeys(maxed_out['hashtag'], None)
for ht in hashtag_posts.keys():
    posts = [cattownposts['id'][i] for i in cattownposts.index if ht in cattownposts['hashtags'][i]]
    hashtag_posts[ht] = posts
hashtag_posts_df = pd.DataFrame.from_dict(hashtag_posts, orient='columns')
print(hashtag_posts_df)
```
