```python
import os
from dotenv import find_dotenv
import pandas as pd
from datetime import datetime
import ast
```

```python
project_dir = find_dotenv().split('.')[0]
data_dir = os.path.join(project_dir, 'data')
raw_data_dir = os.path.join(data_dir, 'raw')
processed_data_dir = os.path.join(data_dir, 'processed')
fig_dir = os.path.join(project_dir, 'figures')
os.chdir(project_dir)
```

```python
posts_path = os.path.join(processed_data_dir, 'cattownposts.csv')
posts = pd.read_csv(posts_path)

# Convert string representations to appropriate object type
for col in ['hashtags', 'mentions', 'pic_dim']:
    posts[col] = posts[col].apply(lambda x: ast.literal_eval(x))

for col in ['timestamp', 'date']:
    posts[col] = pd.to_datetime(posts[col])

# Count words in captions, hashtags, and mentions
posts['caption'].fillna('', inplace=True)
posts['word_count'] = posts['caption'].str.split().str.len()
posts['hashtag_count'] = posts['hashtags'].str.len()
posts['mention_count'] = posts['mentions'].str.len()

posts.head(10)
```
